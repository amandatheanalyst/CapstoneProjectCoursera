Problem Statement: The city of Toronto has an aging population. 60% of it's population will soon reach 65. 
It is with that in mind Toronto seeks to ensure it's citizens have enough parks to get fresh air and stretch their legs in. We will look at the existing park locations to identifiy neighborhoods that could improve by having additional parks.

Approach: We will be using Latitude and Longitude data obtained from wikipidea Toronto page in order to map out the existing park locations using a Foursquare query on 'Venues' in that city. Being able to geographically overlay that data will expose any deficits in park placement.
